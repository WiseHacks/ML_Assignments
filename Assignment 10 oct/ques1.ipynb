```python
# Press ctrl + f9 to run
# importing libraries for plotting the graph
import matplotlib.pyplot as plt
# importing scikit learn libraries
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as MDA
from sklearn.decomposition import PCA
```

```python
# loading iris flower dataset 
iris = datasets.load_iris()
x = iris.data[:, [1,2]] 
y = iris.target
```

```python
def plot(xlabel, ylabel, x, y):
  plt.figure(2, figsize = (6,5))
  plt.scatter(x[:,0],x[:,1],c = y,edgecolors='#000000')
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
```

```python
#  Method 1
# Plot the points
plot('width of sepal', 'length of petal', x, y)
```

```python
# Method 2
# Plot the points
new_x = MDA().fit_transform(iris.data,iris.target)
plot('mda feature 1', 'mda feature 2', new_x, y)
```
